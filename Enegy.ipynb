```python
# Import necessary libraries
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# 1. Load and Preprocess Data
# Load the dataset
df = pd.read_csv('owid-energy-data.csv')

# Check if 'year' column exists
print("Columns in dataset:", df.columns.tolist())

# Filter for Iran, Germany, and Japan, and years 2000-2023
countries = ['Iran', 'Germany', 'Japan']
if 'year' not in df.columns:
    raise KeyError("Column 'year' not found in dataset. Check column names and file content.")
df = df[df['country'].isin(countries) & (df['year'] >= 2000) & (df['year'] <= 2023)]

# Handle missing values: Fill with interpolation for numerical columns, or drop if too sparse
numeric_cols = ['renewables_share_energy', 'energy_per_gdp', 'fossil_share_energy', 
                'population', 'gdp', 'solar_consumption', 'wind_consumption', 
                'hydro_consumption', 'renewables_consumption']
df[numeric_cols] = df[numeric_cols].interpolate(method='linear', limit_direction='both')
df = df.dropna(subset=['renewables_share_energy', 'energy_per_gdp', 'fossil_share_energy'])

# Print data summary to check
print("Data after filtering and preprocessing:")
print(df[['country', 'year', 'renewables_share_energy', 'energy_per_gdp']].head())

# 2. Time Series Analysis with ARIMA
# Function to fit ARIMA and forecast
def arima_forecast(country_df, column, country_name, forecast_years=7):
    series = country_df[column].dropna()
    if len(series) < 10:  # Check if enough data for ARIMA
        print(f"Insufficient data for ARIMA on {column} in {country_name}")
        return None, None
    try:
        # Fit ARIMA model (p=1, d=1, q=1) - simple starting point
        model = ARIMA(series, order=(1, 1, 1))
        model_fit = model.fit()
        # Forecast for 2024-2030
        forecast = model_fit.forecast(steps=forecast_years)
        forecast_index = range(2024, 2024 + forecast_years)
        # Plot
        plt.figure(figsize=(10, 6))
        plt.plot(series.index, series, label='Historical', color='#1f77b4')
        plt.plot(forecast_index, forecast, label='Forecast', color='#ff7f0e', linestyle='--')
        plt.title(f'{column} Trend and Forecast for {country_name} (2000-2030)')
        plt.xlabel('Year')
        plt.ylabel(column)
        plt.legend()
        plt.grid(True)
        plt.savefig(f'{country_name}_{column}_arima.png')
        plt.show()
        return model_fit, forecast
    except Exception as e:
        print(f"ARIMA failed for {column} in {country_name}: {e}")
        return None, None

# Apply ARIMA for each country and key metrics
for country in countries:
    country_df = df[df['country'] == country].set_index('year')
    for col in ['renewables_share_energy', 'energy_per_gdp']:
        arima_forecast(country_df, col, country)

# 3. Clustering with K-Means
# Prepare data for clustering
cluster_features = ['renewables_share_energy', 'fossil_share_energy', 'energy_per_capita', 'gdp']
cluster_df = df[df['year'] == 2023][['country'] + cluster_features].dropna()

# Scale features
scaler = StandardScaler()
cluster_scaled = scaler.fit_transform(cluster_df[cluster_features])

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
cluster_df['Cluster'] = kmeans.fit_predict(cluster_scaled)

# Plot clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=cluster_df, x='renewables_share_energy', y='fossil_share_energy', 
                hue='Cluster', style='country', palette='bright', s=100)
plt.title('Clustering of Iran, Germany, and Japan (2023)')
plt.xlabel('Renewables Share of Energy (%)')
plt.ylabel('Fossil Share of Energy (%)')
plt.legend()
plt.grid(True)
plt.savefig('clustering_2023.png')
plt.show()
print("Cluster assignments (2023):")
print(cluster_df[['country', 'Cluster']])

# 4. Predictive Modeling with Random Forest
# Prepare data for prediction
features = ['population', 'gdp', 'solar_consumption', 'wind_consumption', 'hydro_consumption']
target = 'renewables_share_energy'
# Include 'year' in the selection to use for splitting
model_df = df[['year'] + features + [target]].dropna()

# Print data summary to check
print("Data for Random Forest modeling:")
print(model_df.head())

# Split data (simple split: last 4 years for testing)
train = model_df[model_df['year'] < 2020]
test = model_df[model_df['year'] >= 2020]
X_train = train[features]
y_train = train[target]
X_test = test[features]
y_test = test[target]

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train Random Forest
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train_scaled, y_train)

# Predict and evaluate
y_pred = rf.predict(X_test_scaled)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Random Forest Model Performance:")
print(f"Mean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.2f}")

# Feature importance
feature_importance = pd.DataFrame({'Feature': features, 'Importance': rf.feature_importances_})
feature_importance = feature_importance.sort_values('Importance', ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(data=feature_importance, x='Importance', y='Feature', color='#1f77b4')
plt.title('Feature Importance for Renewables Share Prediction')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.savefig('feature_importance.png')
plt.show()

# 5. Save Results
# Save processed data for further analysis
df.to_csv('processed_energy_data.csv', index=False)
print("Processed data saved to 'processed_energy_data.csv'")
```
